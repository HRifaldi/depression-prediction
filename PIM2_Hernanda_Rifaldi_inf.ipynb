# Import Libraries

```python
import re
from pathlib import Path

import numpy as np
import pandas as pd
import joblib
```

# Loading Model

```python
model_path = Path("artifacts") / "best_model.joblib"
best_model = joblib.load(model_path)

print("Loaded model from:", model_path.resolve())
```

```python
df_raw = pd.read_csv("student_depression_dataset.csv")

def to_snake(s: str) -> str:
    s = s.strip().lower()
    s = s.replace("/", " ")
    s = s.replace("?", "")
    s = re.sub(r"[^0-9a-zA-Z]+", "_", s)
    s = re.sub(r"_+", "_", s).strip("_")
    return s

def clean_dataset(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]

    for c in df.select_dtypes(include="object").columns:
        df[c] = df[c].astype(str).str.strip().str.strip("'").str.strip()

    if "Financial Stress" in df.columns:
        df["Financial Stress"] = df["Financial Stress"].replace("?", np.nan).astype(float)

    df = df.rename(columns={c: to_snake(c) for c in df.columns})
    df = df.rename(columns={
        "have_you_ever_had_suicidal_thoughts": "suicidal_thoughts",
        "family_history_of_mental_illness": "family_history_mental_illness",
    })
    return df

df = clean_dataset(df_raw)


new_data = df.sample(10, random_state=42).drop(columns=["depression"], errors="ignore")
new_data = new_data.drop(columns=["id"], errors="ignore")

display(new_data.head())
```

# Model Inference Prediction

```python
pred = best_model.predict(new_data)

result = new_data.copy()
result["predicted_depression"] = pred

display(result)
```
